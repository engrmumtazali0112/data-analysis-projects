```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_56.csv'  # Adjust to your actual file path

# Load the dataset
try:
    # Skip the first 4 rows of metadata and load the data
    data = pd.read_csv(file_path, skiprows=4, delimiter=',')
    print("File loaded successfully!")
except Exception as e:
    print("Error loading file:", e)
    raise

# Ensure the dataset is properly loaded before proceeding
if 'data' in locals() and not data.empty:
    # Select the relevant year for analysis (e.g., 2021)
    year = "2021"

    # Filter the data for the selected year and drop missing values
    filtered_data = data[['Country Name', year]].dropna()

    # Rename columns for clarity
    filtered_data.rename(columns={year: 'Population'}, inplace=True)

    # Sort data for visualization
    sorted_data = filtered_data.sort_values(by='Population', ascending=False).head(20)  # Top 20 countries

    # Plot a bar chart for population distribution
    plt.figure(figsize=(12, 6))
    plt.bar(sorted_data['Country Name'], sorted_data['Population'], color='skyblue')
    plt.xticks(rotation=45, ha='right')
    plt.title(f"Population Distribution Across Top 20 Countries ({year})", fontsize=14)
    plt.xlabel("Country", fontsize=12)
    plt.ylabel("Population", fontsize=12)
    plt.tight_layout()
    plt.show()
else:
    print("The dataset could not be loaded. Please check the file path or format.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = 'Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_56.csv'

# Load the dataset
try:
    data = pd.read_csv(file_path)
    print("File loaded successfully!")
except Exception as e:
    print("Error loading file:", e)
    raise

# Ensure the dataset is properly loaded
if 'data' in locals() and not data.empty:
    # Inspect and use relevant columns
    if 'Region' in data.columns and 'IncomeGroup' in data.columns:
        # Drop rows with missing values in these columns
        filtered_data = data[['Region', 'IncomeGroup']].dropna()

        # Group by 'Region' to visualize the distribution of countries in different regions
        region_distribution = filtered_data['Region'].value_counts()

        # Plot a bar chart for region distribution
        plt.figure(figsize=(10, 6))
        plt.bar(region_distribution.index, region_distribution.values, color='skyblue')
        plt.xticks(rotation=45, ha='right')
        plt.title("Distribution of Countries by Region", fontsize=14)
        plt.xlabel("Region", fontsize=12)
        plt.ylabel("Number of Countries", fontsize=12)
        plt.tight_layout()
        plt.show()
    else:
        print("Required columns ('Region', 'IncomeGroup') are not in the dataset.")
else:
    print("The dataset could not be loaded. Please check the file path or format.")
```
